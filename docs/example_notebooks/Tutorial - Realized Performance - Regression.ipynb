```python
import nannyml as nml
from IPython.display import display

reference_df = nml.load_synthetic_car_price_dataset()[0]
analysis_df = nml.load_synthetic_car_price_dataset()[1]
analysis_target_df = nml.load_synthetic_car_price_dataset()[2]
analysis_df = analysis_df.join(analysis_target_df)

display(reference_df.head(3))
```

```python
print(reference_df.head(3).to_markdown(tablefmt="grid"))
```

```python
calc = nml.PerformanceCalculator(
    y_pred='y_pred',
    y_true='y_true',
    timestamp_column_name='timestamp',
    problem_type='regression',
    metrics=['mae', 'mape', 'mse', 'msle', 'rmse', 'rmsle'],
    chunk_size=6000)

calc.fit(reference_df)
```

```python
results = calc.calculate(analysis_df)
display(results.filter(period='analysis').to_df())
```

```python
print(results.filter(period='analysis').to_df().to_markdown(tablefmt="grid"))
```

```python
display(results.filter(period='reference').to_df())
```

```python
print(results.filter(period='reference').to_df().to_markdown(tablefmt="grid"))
```

```python
figure = results.plot(kind='performance')
figure.show()
```

```python
figure.write_image(f'../_static/tutorials/performance_calculation/regression/tutorial-performance-calculation-regression.svg')
```

```python
figure = results.filter(period='all', metrics=['rmse']).plot(kind='performance')
figure.show()
figure.write_image(f'../_static/tutorials/performance_calculation/regression/tutorial-performance-calculation-regression-RMSE.svg')
```
